```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
web_graph = nx.DiGraph()

# Add nodes (web pages)
web_graph.add_nodes_from(["Page A", "Page B", "Page C", "Page D"])

# Add edges (links between pages)
web_graph.add_edges_from([
    ("Page A", "Page B"),  # A links to B
    ("Page A", "Page C"),  # A links to C
    ("Page B", "Page C"),  # B links to C
    ("Page C", "Page A"),  # C links to A
    ("Page C", "Page D"),  # C links to D
    ("Page D", "Page B"),  # D links to B
])

# Calculate PageRank
page_rank = nx.pagerank(web_graph)
print("PageRank Scores:")
for page, score in page_rank.items():
    print(f"{page}: {score:.4f}")

# Visualize the web graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(web_graph)  # Position nodes
nx.draw(web_graph, pos, with_labels=True, node_size=2000, node_color="lightblue", font_size=10, font_weight="bold")
plt.title("Web Structure Graph")
plt.show()
```
